```python
from collections import Counter
import matplotlib.pyplot as plt
%matplotlib notebook
import numpy as np
import pandas as pd
```

```python
pai = []

maximum = 2500
for n in range(100, maximum):
    x = np.random.rand(n)
    y = np.random.rand(n)

    zipped = zip(x, y)
    m = [x ** 2 + y ** 2 <= 1 for (x, y) in zipped]

    inside = Counter(m)
    S = inside[True]/n
    _pai = 4 * S
    pai.append(_pai)
    
N = list(range(100, maximum))

plt.figure()
plt.plot(N, pai)
plt.plot(N, [np.pi] * (maximum - 100))
```
